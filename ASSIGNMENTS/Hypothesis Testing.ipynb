# 1.Hypothesis Testing

### 1. State the Hypotheses statement:

###### Null Hypothesis(H0): The mean weekly operating cost is equal to 4000
###### Alternate Hypothesis(Ha) : The mean weekly operating cost is greater than 4000

### 2. calculate the test statistic

```python
import math

x_bar = 3050
mu = 4000 # W = 1000 + 5*600 = 400
sigma = 5*25
n = 25

# calculate standard error
SE = sigma/math.sqrt(n)

# calculate Test statistic
t = (x_bar - mu )/ SE

t
```

### Determine the critical Value

alpha = 5% (0.05)
Area to the left = 1- 0.05 =0.95 
z-table value closest to 0.9500 corresponds to 1.645.
Z-critical = 1.645

### Make a Decision

t statistic = -37
z critical = 1.645

Decision Rule: Reject H0 if t > z

Don't reject Null Hypothesis 

### 5.Conclusion:

Since the calculated test statistic (-38) is less than the critical value (1.645), we do not reject the null hypothesis.
There is no evidence to support the claim that the weekly operating costs have increased.
