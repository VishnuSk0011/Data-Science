## Prime Numbers:

```python
num = int(input("Enter a number :"))

if num > 1:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

## Hospital Bill System

```python
# Hospital_charge_details
Room_charge_per_day = 2000
Doctor_consultation_fee =1500
Lab_test_charges = 300
Discount_threshold = 10000
Discount_rate = 0.10

# user_details
Day_stayed = int(input("Number of days a person has stayed : "))
Num_lab_tests = int(input("Total number of test taken : "))
Medicine_charge = int(input("Total cost of medicine provided : "))

# Calculate charges
Room_charges = Room_charge_per_day * Day_stayed
Lab_charges = Num_lab_tests * Lab_test_charges

# Discount
Total_before_discount = Room_charges + Lab_charges + Doctor_consultation_fee + Medicine_charge

if Total_before_discount > Discount_threshold:
    discount = Total_before_discount * Discount_rate
else:
    discount = 0

# total bill
final_bill = Total_before_discount - discount

print("------ Hospital Billing System ------")
print(f"Room charges : {Room_charges}")
print(f"Lab charges : {Lab_charges}")
print(f"Total before discount : {Total_before_discount}")
print("-----------------------------")
print(f"final bill : {final_bill}")
```

## Word Counter

```python
input_text = input("Enter text here: ")

words = input_text.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
for word, count in word_count.items():
    print(f"'{word}' : {count}")
```

## check for palindrome

```python
def is_palindrome(text):
    text = text.lower()

    cleaned_text = ""
    for char in text:
        if char.isalnum():
            cleaned_text += char

    return cleaned_text == cleaned_text[::-1]

user_input = input("Enter a word or Phrse: ")
if is_palindrome(user_input):
    print("True")
else:
    print("False")
```
