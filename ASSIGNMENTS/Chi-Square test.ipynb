## 1. State the Hypotheses:

Null Hypothesis : There is no association between device type and customer satisfaction.
Alternative Hypothesis : There is an association between device type and customer satisfaction.

## 2. Compute the Chi-Square Statistic:


```python
import math

observed = [
    [50, 70],
    [80, 100],
    [60, 90],
    [30, 50],
    [20, 50]
]

#Row totals, column totals, and grand total

row_totals = [sum(row) for row in observed]
col_totals = [sum(row[0] for row in observed), sum(row[1] for row in observed)]
grand_total = sum(row_totals)

#Calculate Chi-Square
chi_square = 0

for i in range(len(observed)):
    for j in range(len(observed[0])):
        expected = (row_totals[i] * col_totals[j]) / grand_total
        chi_square += (observed[i][j] - expected) ** 2 / expected

chi_square

```

## 3. Determine the Critical Value

alpha = 0.05
degree of freedom(df) = 5-1 = 4
From the Chi-Square distribution table at alpha = 0.05 and df = 4 :

Chi-Square-Critical = 9.488

## 4. Make a Decision:

Decision Rule : Reject Null Hypothesis if Chi_Square Greater than Critical

So , Failed to Reject Null Hypothesis

There is not enough evidence to suggest an association between device type and customer satisfaction.
